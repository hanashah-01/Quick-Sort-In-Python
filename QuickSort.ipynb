String Sorting Algorithm: Quick Sort

```python
import time
start_time = time.time()
```

```python
# This method finds the position of the partition
def partition(array, low, high):

  # The rightmost element which is the last element in the array is selected as the pivot
  pivot = array[high]

  # Pointer for the greater element
  i = low - 1

  # Traverse through each element and the element is compared with the pivot
  for j in range(low, high):
    # If the element is equal or smaller than the pivot
    if array[j] <= pivot:
      # Swap the element smaller than pivot with the greater element pointed by i
      i = i + 1

      # Swap element at i with element at j
      (array[i], array[j]) = (array[j], array[i])

  # Swap the pivot element with the greater element specified by i
  (array[i + 1], array[high]) = (array[high], array[i + 1])

  # return the position of the pivot from where partition is done
  return i + 1
```

```python
# Method to perform quicksort
# Pass the array containing the word list and the first and last element of the word list
def quickSort(array, low, high):
  if low < high:

    # Find the position of the pivot where all elements on the left of the pivot are smaller
    # and all elements on the right of the pivot are greater
    pivot = partition(array, low, high)

    # Recursive call on the left of the pivot
    quickSort(array, low, pivot - 1)

    # Recursive call on the right of the pivot
    quickSort(array, pivot + 1, high)
```

```python
#Read all lines in the txt file and assign it to an array
with open("wordList.txt") as wordList:
    string = wordList.readlines()
```

```python
#Get the number of elements in the array
size = len(string)

#Call the function
quickSort(string, 0, size - 1)
```
